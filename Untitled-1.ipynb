Para poder hacer los diferentes cruces de información entre las diferentes tablas se requiere crear una llave única la cual permitirá cruzar de manera correcta los diferentes datosa, en este caso, se concatenar el número de la consulta y el número de identificación del deudor, ya que son datos únicos, lo que permite la consistencia al momento de cualquier cruce de información.

***Pivoteo de tablas***
Según la estructura que se identifica de los datos remitidos se hará un pivoteo de tabla, donde se tomara como indices el número de la consulta junto al número de cédula (identificación unica para cada cliente). Por otro lado, se trasladará a una columna por separado el tipo de reporte y se le asginaran los valores pertinentes.

```python
# Pivotar el DataFrame para que 'regpaquet' se convierta en columnas
# Usaremos 'regconsul' y 'regcedula' como índices (identificación única para cada cliente)
# Elegimos 'regtotsal' como un ejemplo del valor a pivotar, puedes cambiar esto según tus necesidades

cifin_pivot = cifin_registro.pivot_table(index=['regconsul', 'regcedula'], 
                                         columns='regpaquet', 
                                         values=['regtotsal', 'regnumobl','regobldia','regsaldia','regcuodia','regoblmor','regsalmor','regcuomor'], 
                                         aggfunc='first').reset_index()
```

Al pivotear una tabla, el nombre de las columnas terminará siendo el mismo para todos los casos, es decir, todos los valores quedan con las etiquetas del tipo de reporte, por lo tanto se concatena, los labels del tipo reporte junto al de los valores que fueron trasladados de columnas a filas.

Ya que para el análisis es relevante tener el número de cuentas bancarias que tiene el cliente, principalmente validar aquellas que se encuentran embargadas, se opta por realizar un agrupamiento con el número de consulta y el número de cédula donde se cuente el número de veces que aparece una misma consutla, para así determinar cuantas cuentas bancarias tiene un cliente. **CARGUE DE LA TASA DE USURA E INFLACION POR PERIODO**

Para el análisis pertinente se tiene en cuenta la tasa de usura del periodo en el cual se realizó el crédito. Por lo tanto se realiza el cargue del historico por mes de la tasa limite de usura y se cruza contra la base general de clientes por medio de la fecha de desembolso.

```python
tasa_usura = pd.read_excel("D:/DATOS/Downloads/Tasa de usura.xlsx")
tasa_usura['Fecha inicio'] = tasa_usura['Fecha inicio'].dt.date
tasa_usura['Fecha fin'] = tasa_usura['Fecha fin'].dt.date
```

```python
# Realizamos un merge condicional basándonos en los rangos de fechas
# Crear una función que verificará si la fecha de desembolso cae dentro del rango de fechas en tasa_usura
def obtener_tasa(fecha_desembolso, tasa_usura):
    # Filtrar donde la fecha de desembolso esté dentro del rango
    tasa = tasa_usura[(tasa_usura['Fecha inicio'] <= fecha_desembolso) & 
                      (tasa_usura['Fecha fin'] >= fecha_desembolso)]
    
    # Devolver múltiples columnas si se encuentra una coincidencia
    if not tasa.empty:
        return pd.Series([tasa['EA'].values[0], tasa['Inflación total 1'].values[0]])  # Devuelve varias columnas
    else:
        return pd.Series([pd.NA, pd.NA])  # Devuelve NaN si no hay coincidencia

# Aplicar la función a la columna 'FECHA DESEMBOLSO' de base_clientes
base_clientes[['TASA_USURA', 'Inflación total 1']] = base_clientes['FECHA DESEMBOLSO'].apply(obtener_tasa, tasa_usura=tasa_usura)
```

```python
# Ver el tipo de datos de las columnas 'TASA_USURA' e 'Inflación total 1'
base_clientes['TASA_USURA'] = pd.to_numeric(base_clientes['TASA_USURA'], errors='coerce')
base_clientes['Inflación total 1'] = pd.to_numeric(base_clientes['Inflación total 1'], errors='coerce')
print(base_clientes[['TASA_USURA', 'Inflación total 1']].dtypes)
```

**Falta de variabilidad**

 Un alto porcentaje de ceros indica que la variable tiene poca variación. Los modelos estadísticos y de aprendizaje automático dependen de la variabilidad en los datos para identificar patrones y relaciones. La ausencia de variabilidad limita la capacidad del modelo para aprender y generalizar, lo que puede conducir a resultados poco precisos o sesgados. Por lo tanto se deciden excluir las variables que tengan mas del 70% de ceros.

```python

# Seleccionar solo las columnas numéricas del DataFrame
numeric_columns = base_clientes.select_dtypes(include=['float64', 'int64'])

# Contar los ceros en cada columna numérica
zeros_count = (numeric_columns == 0).sum()

# Calcular el porcentaje de ceros por columna
total_rows = len(base_clientes)  # Número total de filas en el DataFrame
zeros_percentage = (zeros_count / total_rows) * 100

# Crear un DataFrame para mostrar los resultados
zeros_summary = pd.DataFrame({
    'zero_count': zeros_count,
    'zero_percentage': zeros_percentage
})
```

```python
columnas_cualitativas = base_clientes1.select_dtypes(include=['object', 'category'])

# Contar el número de nulos en cada columna cualitativa
nulos_cualitativos = columnas_cualitativas.isnull().sum()

# Calcular el porcentaje de nulos en cada columna cualitativa
porcentaje_nulos = (nulos_cualitativos / len(base_clientes1)) * 100

# Combinar ambos resultados en un DataFrame
resultado_nulos = pd.DataFrame({'Nulos': nulos_cualitativos, 'Porcentaje Nulos': porcentaje_nulos})
```

```python
# Reemplazar nulos por la moda de cada columna
for columna in columnas_cualitativas.columns:
    moda = base_clientes1[columna].mode()[0]  # Obtener la moda (el valor más frecuente)
    base_clientes1[columna] = base_clientes1[columna].fillna(moda)  # Reemplazar los nulos con la moda sin usar inplace
```

```python

# Función para eliminar tildes y convertir a mayúsculas
def normalizar_texto(texto):
    if isinstance(texto, str):  # Solo procesar si es string
        texto = texto.upper()  # Convertir a mayúsculas
        # Eliminar tildes
        texto = ''.join((c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn'))
    return texto

# Seleccionar solo las columnas cualitativas (categóricas)
columnas_cualitativas = base_clientes1.select_dtypes(include=['object', 'category'])

# Aplicar la normalización a todas las columnas categóricas
for columna in columnas_cualitativas.columns:
    base_clientes1[columna] = base_clientes1[columna].apply(normalizar_texto)
```
